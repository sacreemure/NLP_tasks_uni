```python
from bs4 import BeautifulSoup
import re

with open('/content/antipatija.man.contexts.txt', 'r', encoding='windows-1251') as f:
  data = f.read()

data = BeautifulSoup(data, 'lxml').find_all('se')

# поиск по лемме,
def search_lemma(lemma, data):
    with open('lemma.txt', 'w') as f:
        lemma = lemma.lower().strip()
        contexts = []
        frequency = 0
        for sent in data:
            if re.search(f'lex="{lemma}"', str(sent('w'))):
                frequency += 1
                context = ' '.join([word.text for word in sent('w')])
                contexts.append(context)
        f.write(f"Лемма: {lemma} \nКонтексты:\n")
        for context in contexts:
            f.write(context+'\n')
        f.write(f"Частотa вхождения: {frequency}\n")


# поиск по тегу,
def search_tag(tag, data):
    with open('tag.txt', 'w') as f:
        frequency = 0
        tag = tag.lower().strip()
        f.write(f"Тег: {tag}\n")
        f.write("Контексты:\n")
        for sent in data:
            for word in sent('w'):
                try:
                    if tag in str(word.ana['sem']).strip().split():
                        frequency += 1
                        context = ' '.join([word.text for word in sent('w')])
                        f.write(f'{context}, тег "{tag}", словоформа "{word.text}"\n')
                except KeyError:
                    pass
        f.write(f"Частотa вхождения: {frequency}\n")


# поиск по лемме и тегу,
def search_lemma_and_tag(lemma, tag, data):
    with open('lemma_and_tag.txt', 'w') as f:
        frequency = 0
        lemma = lemma.lower().strip()
        tag = tag.lower().strip()
        f.write(f"Лемма: {lemma}, тег: {tag}\n")
        f.write("Контексты:\n")
        for sent in data:
            for word in sent('w'):
                try:
                    if word.ana['lex'] == lemma and tag.lower() in str(word.ana['sem']).strip().split():
                        frequency += 1
                        context = ' '.join([word.text for word in sent('w')])
                        f.write(f'{context}, лемма "{lemma}", тег "{tag}"\n')
                except KeyError:
                    pass
        f.write(f"Частотa вхождения: {frequency}\n")

# поиск по комбинации тегов,
def search_tags(tag1, tag2, data):
    with open('tags.txt', 'w') as f:
        frequency = 0
        tag1 = tag1.lower().strip()
        tag2 = tag2.lower().strip()
        f.write(f"Тег 1: {tag1}, тег 2: {tag2}\n")
        f.write("Контексты:\n")
        for sent in data:
            for word in sent('w'):
                try:
                    sem = str(word.ana['sem']).strip().split()
                    if tag1 in sem and tag2 in sem:
                        frequency += 1
                        lemma = str(word.ana['lex']).strip()
                        context = ' '.join([word.text for word in sent('w')])
                        f.write(f'{context}, лемма "{lemma}"\n')
                except KeyError:
                    pass
        f.write(f"Частотa вхождения: {frequency}\n")

#Произвести тестирование на фрагменте корпуса (см. выше).

search_lemma(input('Введите лемму: '), data)
search_tag(input('Введите тег: '), data)
inp = input('Введите лемму и тег через пробел: ').split()
lemma = inp[0]
tag = inp[1]
search_lemma_and_tag(lemma, tag, data)
inp = input('Введите теги через пробел: ').split()
tag1 = inp[0]
tag2 = inp[1]
search_tags(tag1, tag2, data)
```
