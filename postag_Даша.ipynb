```python
import pymorphy2
import nltk
#nltk.download('punkt')

from nltk.tokenize import word_tokenize
morph1 = pymorphy2.MorphAnalyzer()

list1 = []

with open('rueval_2010_goldstandard_text.txt', 'r', encoding='Windows-1251') as m:
          for line in m:
            line = word_tokenize(str(line))
            for word in line:
              list1.append((word, morph1.parse(word)[0].tag.POS))

print(list1)
```

```python
import json
from pymystem3 import Mystem


with open("/content/rueval_2010_goldstandard_text.txt", "r", encoding="Windows-1251") as f:
  text = f.read()
  m = Mystem()
  lemmas = m.lemmatize(text)

print("lemmas:", ''.join(lemmas))
print("full info:", json.dumps(m.analyze(text), ensure_ascii=False))
```

**Stanza**

```python
!pip install stanza
```

```python
import stanza
stanza.download('ru')
nlp = stanza.Pipeline('ru')
```

```python
list4 = []

with open("rueval_2010_goldstandard_text.txt", "r", encoding="Windows-1251") as m:
  for line in m:
    doc = nlp(line)
    print(*[f'word: {word.text}\tlemma: {word.lemma}\tupos: {word.upos}\tfeats: {word.feats if word.feats else "_"}' for sent in doc.sentences for word in sent.words], sep='\n')
    for sentence in doc.sentences:
      for word in sentence.words:
        list4.append((word.text, word.upos))

print(list4)
```

**RNNmorph**

```python
!pip install rnnmorph
```

```python
from rnnmorph.predictor import RNNMorphPredictor
predictor = RNNMorphPredictor(language='ru')

list5 = []

with open("rueval_2010_goldstandard_text.txt", "r", encoding="Windows-1251") as m:
  for line in m:
    line = word_tokenize(str(line))
    forms = predictor.predict(line)
    for form in forms:
      list5.append((form.word, form.pos))

print(list5)
```
